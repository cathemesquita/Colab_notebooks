$ D = \sqrt{(x_B - x_A)² + (y_B- y_A)²} $

* 1) definir os pontos como tupla: A = (x_A,y_A), B = (x_B,y_B)
* 2) para acessar os pontos da tupla: A[0] é o x de A, A[1] é o y de A

> Teorema de Pitágoras:

 $ hip² = co² + ca² $
